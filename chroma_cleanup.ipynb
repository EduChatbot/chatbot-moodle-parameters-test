```python
import chromadb
from chromadb.config import Settings

# Konfiguracja połączenia (zgodna z Twoim settings.py)
CHROMA_HOST = "84.8.152.23"
CHROMA_PORT = 8000
```

```python
def print_chromadb():
    print(f"Łączenie z ChromaDB na {CHROMA_HOST}:{CHROMA_PORT}...")
    client = chromadb.HttpClient(host=CHROMA_HOST, port=CHROMA_PORT)
        
        # Pobranie listy wszystkich kolekcji
    collections = client.list_collections()
        
    if not collections:
        print("Baza jest pusta.")
        return

    print(f"Znaleziono {len(collections)} kolekcji.")

    for collection in collections:
        print(f"{collection.name}")
```

```python
print_chromadb()
```

```python
def delete_collection_by_name(collection_name: str):
    print(f"Łączenie z ChromaDB na {CHROMA_HOST}:{CHROMA_PORT}...")

    try:
        client = chromadb.HttpClient(
            host=CHROMA_HOST,
            port=CHROMA_PORT
        )

        print(f"Usuwanie kolekcji: {collection_name}")
        client.delete_collection(name=collection_name)

        print("✅ Kolekcja została usunięta.")

    except Exception as e:
        print(f"❌ Błąd: {e}")
```

```python
delete_collection_by_name("course_11_voyage_large_2_2000_400")
```
